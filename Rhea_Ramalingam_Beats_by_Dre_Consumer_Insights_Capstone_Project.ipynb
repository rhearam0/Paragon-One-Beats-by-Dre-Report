## 5. Exploratory Data Analysis (EDA)

### 5.1 Data Overview
Looking at the dataset, I was able to get the mean, median, and mode and standard deviation from the ratings of the reviews. So this meant the average of all the ratings (5 stars being the highest). As seen in the code below:  

```python
#Explatory Data Analysis
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df_reviews = pd.read_csv('/content/merged_data.csv')

# Calculate descriptive statistics
mean_rating = df_reviews['rating'].mean()
median_rating = df_reviews['rating'].median()
mode_rating = df_reviews['rating'].mode()[0]
variance_rating = df_reviews['rating'].var()
std_rating = df_reviews['rating'].std()
quantiles_rating = df_reviews['rating'].quantile([0.25, 0.5, 0.75])

print("Mean Rating:", mean_rating)
print("Median Rating:", median_rating)
print("Mode Rating:", mode_rating)
print("Variance Rating:", variance_rating)
print("Standard Deviation Rating:", std_rating)
print("Quantiles Rating:", quantiles_rating)
```

The analysis of the product ratings reveals several key insights into customer satisfaction. First, the mean rating of 4.78 out of 5 indicates that customers are generally very satisfied with the product, reflecting a strong positive reception. Second, both the median and mode ratings are 5 stars, emphasizing that the majority of customers rated the product at the highest possible level, further reinforcing the overall positive sentiment. Third, the standard deviation of 0.50 suggests that there is some variability in the ratings, but not significant enough to undermine the overall high satisfaction. Finally, the variance of 0.25 indicates that while most customers gave high ratings, there is a moderate spread in the scores, though this does not detract significantly from the overwhelmingly positive trend. Overall, the data suggests that the product is very well-received, with most customers experiencing high levels of satisfaction.

## 6. Sentiment Analysis

### 6.1 Methodology
Below are the graphs created from several reviews of the Beats Speaker product to analyze the sentiment levels.

```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
file_path = '/content/merged_data.csv'
data = pd.read_csv(file_path)

# Filter for Beats by Dre products
beats_data = data[data['title'].str.contains('Beats', case=False, na=False)]

# Perform sentiment analysis
def get_sentiment(text):
    blob = TextBlob(str(text))
    sentiment_score = blob.sentiment.polarity
    return sentiment_score

beats_data['sentiment_score'] = beats_data['content'].apply(get_sentiment)

def classify_sentiment(score):
    if score > 0:
        return 'Positive'
    elif score < 0:
        return 'Negative'
    else:
        return 'Neutral'

beats_data['sentiment'] = beats_data['sentiment_score'].apply(classify_sentiment)

# Aggregate the sentiment scores
sentiment_counts = beats_data['sentiment'].value_counts()

# Plotting the sentiment classification
plt.figure(figsize=(8, 6))
sns.countplot(x='sentiment', data=beats_data, order=['Positive', 'Neutral', 'Negative'])
plt.title('Sentiment Analysis of Beats by Dre Reviews')
plt.xlabel('Sentiment')
plt.ylabel('Number of Reviews')
plt.show()

# Show the overall sentiment analysis results
print("Sentiment Analysis Results for Beats by Dre:")
print(sentiment_counts)
```

### 8.2 Comparison of Sentiments
Compare the sentiment analysis results of Beats by Dre with its competitors using visualizations to highlight differences and similarities.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob

# Load your dataset
df = pd.read_csv('/content/merged_data.csv')

# Perform sentiment analysis using TextBlob
def get_sentiment(text):
    blob = TextBlob(str(text))
    return blob.sentiment.polarity

df['sentiment_score'] = df['content'].apply(get_sentiment)

# Classify sentiments based on polarity score
def classify_sentiment(score):
    if score > 0.1:
        return 'Positive'
    elif score < -0.1:
        return 'Negative'
    else:
        return 'Neutral'

df['sentiment'] = df['sentiment_score'].apply(classify_sentiment)

# Filter data for each brand by ASIN (product ID)
brand_asin_map = {
    'Beats by Dre': 'B0D4SX9RC6',
    'Sony': 'B0CXL4FQBK',
    'JBL': 'B08X4YMTPM',
    'Bose': 'B099TLMRB6'
}

# **Replace 'product_id' with the actual name of the column containing ASINs in your DataFrame**
asin_column_name = 'product_id'

# Create bar plots for each brand
for brand, asin in brand_asin_map.items():
    brand_data = df[df[asin_column_name] == asin]  # Filter using the correct column name

    # Create a bar plot for the sentiment distribution
    plt.figure(figsize=(8, 6))
    sns.countplot(data=brand_data, x='sentiment', palette='coolwarm')
    plt.title(f'Sentiment Distribution for {brand} (ASIN: {asin})')
    plt.xlabel('Sentiment')
    plt.ylabel('Count of Reviews')
    plt.show()
```

### 9.1 Product Improvements
Based on my research, which included consumer sentiment analysis, reviews, and comparisons with top competitors, I have developed two key recommendations to improve the Beats Pill speaker. The two primary recommendations are as follows:
